### <u>Plot the decision trees</u>

```python
useGraphviz = True
```

```python
import joblib
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree, export_graphviz
import graphviz
import pydotplus

features = [
    ['Pclass','Sex','Age','CoTravelers','Fare','Embarked'],
    ['Pclass','Sex','Age','Fare','Embarked','Relatives'],
    ['Pclass','Sex','Age','Fare','Embarked','Relatives'],
    ['Pclass','Sex','Age','Fare','Embarked','Relatives'],
    ['Pclass','Sex','Age','Fare','Embarked','Relatives'],
    ['Pclass','Sex','Age','Fare','Embarked','Relatives'],
    ['Pclass','Sex','Age','Fare','Embarked','Relatives','IsAlone','Title']
]

# Load the model
for i in range(1, 8):
    model = joblib.load(f"SampleModels/DecisionTree/titanic_decision_tree_model_v{i}.joblib")

    if not useGraphviz:
        plt.figure(figsize=(100, 100))
        plot_tree(model, feature_names = features[i-1], filled=True)
        plt.savefig(f"VisualizedModels/DecisionTree/titanic_decision_tree_model_v{i}_sklearn.png")
        plt.close()

    else:
        dot_data = export_graphviz(model, out_file=None, 
                                feature_names=features[i-1],  
                                filled=True, rounded=True,  
                                special_characters=True)  
        graph = pydotplus.graph_from_dot_data(dot_data)  
        graph.write_png(f"VisualizedModels/DecisionTree/titanic_decision_tree_model_v{i}_graphviz.png")

    print(f"Model v{i} exported")

print("Done")
```
